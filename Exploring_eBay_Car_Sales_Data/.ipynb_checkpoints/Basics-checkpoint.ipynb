### We can see that some of the data has null values, with which we have to do something. Also they use camelcase instead of snakecase which is preferred in Python, so I would need to convert them.

```python
autos.columns
```

```python
autos.rename(columns={'yearOfRegistration': 'registration_year', 
                      'monthOfRegistration': 'registration_month', 
                      'notRepairedDamage': 'unrepaired_damage', 
                      'dateCreated': 'ad_created',
                      'dateCrawled': 'date_crawled',
                     'offerType': 'offer_type',
                     'vehicleType': 'vehicle_type',
                     'powerPS': 'power_PS',
                     'fuelType': 'fuel_type',
                     'nrOfPictures':'nr_of_pictures',
                     'postalCode':'postal_code',
                     'lastSeen' : 'last_seen'}, inplace=True)
```

### I have changed the column names to be snake case and renamed some of the columns to make more sense and so it would be conventional and easier to query the df

```python
autos.describe()
```

### Price and odometer columns are both numberic values but they are stored as text files

```python
autos['price']
```

```python
autos['price'] = autos['price'].str.replace('$' , '')
autos['price'] = autos['price'].str.replace(',' , '')
autos['price'] = autos['price'].astype(int)
```

```python
autos['price']
```

```python
autos['odometer']
```

```python
autos['odometer'] = autos['odometer'].str.replace('km' , '')
autos['odometer'] = autos['odometer'].str.replace(',' , '')
autos['odometer'] = autos['odometer'] .astype(int)
```

```python
autos['odometer']
```

```python
autos.rename(columns = {'odometer' : 'odometer_km', 'price' : 'price_dollars'}, inplace = True)
```

```python
filter_prices_dollars = ~(autos["price_dollars"] <= 150) | ((autos["price_dollars"] >= 1000000) & ~((autos['price_dollars'] == 1300000)
                                         | (autos['price_dollars'] == 3890000)))
```

```python
autos = autos[filter_prices_dollars]
```

### I have now removed all cars that are less than 150 or over 1 000 000 dollars except for 2 Ferraris that had a correct price.
