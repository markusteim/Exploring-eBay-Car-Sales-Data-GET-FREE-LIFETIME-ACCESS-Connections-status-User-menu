### I will use value_counts to get the most popular brands and then take the top 20 most frequent brands to then add them to a dictionary along with their mean price and another dictionary with their mean mileage

```python
top_brands_mean_price = {}


brands = autos['brand'].value_counts().index.tolist()
for car_brand in brands[:20]:
    select_cars = autos[autos['brand'] == car_brand][['brand','price_dollars']]
    
    mean_price = select_cars.iloc[:,1].mean()
    
    top_brands_mean_price[car_brand] = round(mean_price, 2)
    
    
top_brands_mean_mileage = {}


brands = autos['brand'].value_counts().index.tolist()
for car_brand in brands[:20]:
    select_cars = autos[autos['brand'] == car_brand][['brand','odometer_km']]
    
    mean_mileage = select_cars.iloc[:,1].mean()
    
    top_brands_mean_mileage[car_brand] = round(mean_mileage, 2)
```
